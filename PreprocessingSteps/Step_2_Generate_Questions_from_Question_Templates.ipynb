```python
def generate_question_template_1(table_name, disease_name, list_of_symptoms, list_of_signs):
    list_of_questions_template_1 = []
    for symptom_A in list_of_symptoms:
        if (not 'no ' in symptom_A) and (not 'may or may not have' in symptom_A):
            question = 'I have ' + symptom_A + ', what disease do I have?'
            symptom_B, symptom_C, symptom_D, sign_A, sign_B, sign_C, sign_D = '','','','','','',''
            row_entry = [table_name,question,disease_name,symptom_A,symptom_B, symptom_C, symptom_D, sign_A, sign_B, sign_C, sign_D]
            list_of_questions_template_1.append(row_entry)
    return list_of_questions_template_1

list_of_questions_template_1 = generate_question_template_1(table_name,disease_name,list_of_symptoms,list_of_signs)
list_of_questions_template_1
```
